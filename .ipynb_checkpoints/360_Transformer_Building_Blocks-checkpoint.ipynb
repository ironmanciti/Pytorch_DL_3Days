## 1. 환경 설정

````python
import math
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt

torch.manual_seed(42)

# PyTorch 2.6+ 의 torch.accelerator API
if hasattr(torch, "accelerator") and torch.accelerator.is_available():
    device = torch.accelerator.current_accelerator()
else:
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("PyTorch version :", torch.__version__)
print("device          :", device)
````

## 2. 차원 표기 약속

논문/구현마다 다르게 쓰지만, 본 노트북은 일관되게 다음을 사용합니다.

| 기호 | 의미 | 예시 |
|---|---|---|
| `N` | batch size | 2 |
| `L` | 시퀀스 길이 | 8 |
| `D` | 토큰 임베딩 차원 (모델 차원) | 64 |
| `H` | head 개수 | 4 |
| `D_head = D / H` | 각 head 의 차원 | 16 |

핵심 흐름은:  `(N, L, D)` 로 들어와서 → `(N, L, D)` 로 나갑니다. 중간에 `(N, H, L, D_head)` 로 잠깐 head 별로 쪼개졌다가 다시 합쳐집니다.


````python
# 노트북 전체에서 쓸 hyperparameter
N = 2          # batch
L = 8          # 시퀀스 길이
D = 64         # 토큰 차원
H = 4          # head 수
D_head = D // H

print(f"batch={N}, seq_len={L}, dim={D}, n_heads={H}, head_dim={D_head}")

# 더미 입력 (보통은 token embedding + positional encoding 결과가 들어옴)
x = torch.randn(N, L, D, device=device)
x.shape
````

## 3. Scaled Dot-Product Attention — 가장 작은 단위

> **공식**:   `Attention(Q, K, V) = softmax(Q Kᵀ / √d_k) · V`

슬라이드 #125–#127 의 핵심 식 그대로입니다. 한 줄씩 분해하면:

1. **Q · Kᵀ** — 모든 토큰 쌍의 내적 = 유사도 행렬, shape `(L, L)`
2. **/ √d_k** — 스케일링 (큰 d_k 에서 softmax gradient 가 죽는 것을 방지)
3. **softmax** — 한 행을 확률 분포로 변환 → 어느 토큰을 얼마나 볼지 결정
4. **· V** — 그 가중치로 V 들의 가중합 = 새 표현


````python
def scaled_dot_product_attention_manual(q, k, v, mask=None):
    """교육용 — 한 줄씩 풀어서 쓴 SDPA. 실무에서는 F.scaled_dot_product_attention 을 사용한다."""
    # q, k, v shape : (..., L, D_head)
    d_k = q.size(-1)

    # 1) Q · K^T  →  (..., L_q, L_k) 유사도 행렬
    scores = torch.matmul(q, k.transpose(-2, -1)) / math.sqrt(d_k)

    # 2) (선택) mask 적용 — masked 위치는 -inf 로 채워서 softmax 후 0이 되게 함
    if mask is not None:
        scores = scores.masked_fill(mask == 0, float("-inf"))

    # 3) softmax → attention weight
    attn = F.softmax(scores, dim=-1)

    # 4) attention weight · V → 새 표현
    out = torch.matmul(attn, v)
    return out, attn

# 1-head 으로 확인 (head 차원 없이 그냥 (N, L, D))
q = k = v = x      # self-attention 이라 셋이 같음
out, attn = scaled_dot_product_attention_manual(q, k, v)
print("input  :", x.shape)
print("output :", out.shape)
print("attn   :", attn.shape, "  (한 행의 합 ≈ 1.0 이면 softmax OK)")
print("attn[0,0].sum() =", attn[0, 0].sum().item())
````

### 3-1. 어텐션 행렬 시각화

`attn` 의 shape 는 `(N, L_q, L_k)` 이고, **`attn[i, j]` = i 번째 query 토큰이 j 번째 key 토큰에 주는 가중치** 입니다. heatmap 으로 확인합니다 — 학습되지 않은 random weight 라 패턴이 무작위지만, 학습 후에는 의미 있는 구조가 나타납니다.


````python
plt.figure(figsize=(4, 4))
plt.imshow(attn[0].cpu().detach().numpy(), cmap="viridis")
plt.colorbar(label="attention weight")
plt.title("Attention matrix (sample 0)\nrow=query token, col=key token")
plt.xlabel("key idx"); plt.ylabel("query idx")
plt.show()
````

### 3-2. PyTorch 가 제공하는 fused 버전과 비교

실무에서는 직접 짜지 않고 **`F.scaled_dot_product_attention`** 을 씁니다 — 내부적으로 FlashAttention/Memory-Efficient Attention/Math 중 하드웨어에 맞는 fused kernel 로 디스패치되어 훨씬 빠릅니다 (출력 자체는 동일).


````python
out_fused = F.scaled_dot_product_attention(q, k, v)
print("manual vs fused 출력 일치 :", torch.allclose(out, out_fused, atol=1e-6))
````

## 4. Multi-Head Attention

슬라이드 #129 의 핵심: **하나의 attention 으로 보는 것보다, head 를 H 개로 쪼개서 각 head 가 서로 다른 subspace 를 학습** 하면 더 풍부한 관계를 잡아낼 수 있습니다.

흐름:
1. **Input projection** — `(N, L, D)` 입력에 `Linear(D → 3D)` 한 번 곱해서 Q·K·V 한꺼번에 생성 (packed projection — 공식 튜토리얼 권장 패턴)
2. **head split** — `(N, L, D)` → `(N, L, H, D_head)` → `(N, H, L, D_head)`
3. **SDPA** — 각 head 에서 독립적으로 attention
4. **head merge** — `(N, H, L, D_head)` → `(N, L, D)`
5. **Output projection** — `Linear(D → D)`

아래 클래스는 PyTorch 공식 튜토리얼 `transformer_building_blocks.py` 의 `MultiHeadAttention` 을 self-attention 케이스(Q=K=V 의 차원이 같음)로 단순화한 것입니다.


````python
class MultiHeadSelfAttention(nn.Module):
    """공식 튜토리얼 MultiHeadAttention 을 self-attention 용으로 단순화"""

    def __init__(self, dim: int, n_heads: int, dropout: float = 0.0, bias: bool = True):
        super().__init__()
        assert dim % n_heads == 0, "dim 은 n_heads 로 나누어떨어져야 함"
        self.dim = dim
        self.n_heads = n_heads
        self.head_dim = dim // n_heads
        self.dropout = dropout

        # packed projection : Q, K, V 가중치를 하나의 Linear 로 묶음 (3 × dim 출력)
        self.qkv_proj = nn.Linear(dim, 3 * dim, bias=bias)
        self.out_proj = nn.Linear(dim, dim, bias=bias)

    def forward(self, x, is_causal: bool = False, attn_mask=None):
        # x : (N, L, D)
        N, L, D = x.shape

        # 1) packed projection 후 Q/K/V 분리 — 한 번의 matmul 로 끝
        qkv = self.qkv_proj(x)                              # (N, L, 3D)
        q, k, v = torch.chunk(qkv, 3, dim=-1)               # 각 (N, L, D)

        # 2) head split  : (N, L, D) -> (N, L, H, D_head) -> (N, H, L, D_head)
        q = q.unflatten(-1, (self.n_heads, self.head_dim)).transpose(1, 2)
        k = k.unflatten(-1, (self.n_heads, self.head_dim)).transpose(1, 2)
        v = v.unflatten(-1, (self.n_heads, self.head_dim)).transpose(1, 2)

        # 3) SDPA (fused) — H 개 head 에 대해 batch 로 한번에 계산
        attn_out = F.scaled_dot_product_attention(
            q, k, v,
            dropout_p=self.dropout if self.training else 0.0,
            is_causal=is_causal,
            attn_mask=attn_mask,
        )                                                    # (N, H, L, D_head)

        # 4) head merge : (N, H, L, D_head) -> (N, L, H, D_head) -> (N, L, D)
        attn_out = attn_out.transpose(1, 2).flatten(-2)

        # 5) output projection
        return self.out_proj(attn_out)


mha = MultiHeadSelfAttention(dim=D, n_heads=H).to(device)
out = mha(x)
print("input  :", x.shape)
print("output :", out.shape, "  ← 입출력 shape 동일 (residual 가능)")
print("총 파라미터 수 :", sum(p.numel() for p in mha.parameters()))
````

## 5. Positional Encoding

슬라이드 #128 의 메시지: **Self-Attention 은 순서 정보를 모릅니다.** 즉 `[A, B, C]` 와 `[C, B, A]` 가 같은 set 처럼 처리됩니다. 그래서 토큰 임베딩에 **위치별 고유 벡터** 를 더해줍니다.

가장 고전적인 방식 — **Sinusoidal Positional Encoding** :

$$
\mathrm{PE}(pos, 2i) = \sin\left(\frac{pos}{10000^{2i/D}}\right), \quad
\mathrm{PE}(pos, 2i+1) = \cos\left(\frac{pos}{10000^{2i/D}}\right)
$$

장점: 학습 없이 결정론적, **상대 위치(`pos+k`)를 sin/cos 의 선형결합으로 표현** 가능 → 모델이 상대 위치를 자연스럽게 학습.


````python
class SinusoidalPositionalEncoding(nn.Module):
    def __init__(self, dim: int, max_len: int = 5000):
        super().__init__()
        pe = torch.zeros(max_len, dim)                      # (max_len, D)
        position = torch.arange(max_len).unsqueeze(1)       # (max_len, 1)
        div_term = torch.exp(torch.arange(0, dim, 2) * (-math.log(10000.0) / dim))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        # 학습 대상은 아니지만 device 이동에 같이 따라오도록 buffer 로 등록
        self.register_buffer("pe", pe.unsqueeze(0))         # (1, max_len, D)

    def forward(self, x):
        # x : (N, L, D) — 토큰 임베딩 결과
        return x + self.pe[:, : x.size(1)]


pe = SinusoidalPositionalEncoding(dim=D).to(device)
x_pe = pe(x)
print("PE 추가 전후 shape 동일 :", x.shape, "→", x_pe.shape)
````

### 5-1. PE 패턴 시각화 — 왜 이게 "위치"를 표현하나?

````python
# 100 개 위치, 차원 64 의 PE 행렬을 heatmap 으로
pe_matrix = pe.pe.squeeze(0)[:100].cpu().numpy()      # (100, 64)
plt.figure(figsize=(8, 4))
plt.imshow(pe_matrix, aspect="auto", cmap="RdBu_r")
plt.colorbar(label="value")
plt.title("Sinusoidal Positional Encoding")
plt.xlabel("dimension"); plt.ylabel("position")
plt.show()
print("→ 왼쪽 차원은 짧은 주기, 오른쪽 차원은 긴 주기. 모든 위치 벡터가 서로 구별됨.")
````

## 6. Encoder Block — MHA + FFN + LayerNorm + Residual

슬라이드 #124 의 Encoder unit 한 칸은 다음 5가지의 조합입니다.

```
       ┌──────────────────────────┐
  x ──►│  LayerNorm → MHA → +x    │── h
       └──────────────────────────┘
                ▼
       ┌──────────────────────────┐
  h ──►│  LayerNorm → FFN → +h    │── y
       └──────────────────────────┘
```

> 원래 *Attention is All You Need* 는 **Post-LN** (norm 을 sublayer 뒤에) 였지만, 큰 모델 학습 안정성 때문에 GPT/LLaMA 등 현대 모델은 모두 **Pre-LN** (norm 을 sublayer 앞에) 을 씁니다. 본 노트북도 Pre-LN 채택.


````python
class FeedForward(nn.Module):
    """흔한 FFN: Linear → ReLU(또는 GELU) → Linear, 보통 hidden = 4*D"""
    def __init__(self, dim: int, hidden_mult: int = 4, dropout: float = 0.0):
        super().__init__()
        hidden = dim * hidden_mult
        self.fc1 = nn.Linear(dim, hidden)
        self.fc2 = nn.Linear(hidden, dim)
        self.act = nn.GELU()
        self.drop = nn.Dropout(dropout)

    def forward(self, x):
        return self.drop(self.fc2(self.act(self.fc1(x))))


class EncoderBlock(nn.Module):
    def __init__(self, dim: int, n_heads: int, dropout: float = 0.0):
        super().__init__()
        self.norm1 = nn.LayerNorm(dim)
        self.attn  = MultiHeadSelfAttention(dim, n_heads, dropout=dropout)
        self.norm2 = nn.LayerNorm(dim)
        self.ffn   = FeedForward(dim, dropout=dropout)

    def forward(self, x, is_causal=False):
        # Pre-LN + residual
        x = x + self.attn(self.norm1(x), is_causal=is_causal)
        x = x + self.ffn(self.norm2(x))
        return x


block = EncoderBlock(dim=D, n_heads=H).to(device)
print(block)
print("\nblock(x) 출력 shape :", block(x).shape)
````

## 7. Mini Transformer Encoder — 블록 쌓기

논문은 N=6 stack 이지만 학습 데모용으로 4 정도면 충분합니다. 실제로는 token embedding 과 PE 도 함께 묶어 한 모델로 만듭니다.


````python
class TinyEncoder(nn.Module):
    def __init__(self, vocab_size: int, dim: int = 64, n_heads: int = 4,
                 n_layers: int = 4, max_len: int = 256, dropout: float = 0.1):
        super().__init__()
        self.token_emb = nn.Embedding(vocab_size, dim)
        self.pos_enc   = SinusoidalPositionalEncoding(dim, max_len=max_len)
        self.blocks    = nn.ModuleList([
            EncoderBlock(dim, n_heads, dropout=dropout) for _ in range(n_layers)
        ])
        self.norm = nn.LayerNorm(dim)

    def forward(self, token_ids, is_causal=False):
        # token_ids : (N, L) — 정수 토큰 ID
        h = self.token_emb(token_ids)        # (N, L, D)
        h = self.pos_enc(h)                  # (N, L, D)
        for blk in self.blocks:
            h = blk(h, is_causal=is_causal)
        return self.norm(h)


vocab_size = 1000
model = TinyEncoder(vocab_size=vocab_size, dim=D, n_heads=H, n_layers=4).to(device)

token_ids = torch.randint(0, vocab_size, (N, L), device=device)
out = model(token_ids)
print("token_ids :", token_ids.shape)
print("output    :", out.shape)
print("총 파라미터 수 :", f"{sum(p.numel() for p in model.parameters()):,}")
````

## 8. Causal Mask — Encoder vs Decoder/GPT 의 차이

지금까지 만든 건 **양방향(bidirectional) self-attention** = BERT 류 encoder 입니다.

GPT 류 decoder 는 **미래 토큰을 보면 안 되니까** causal(=upper-triangular) mask 를 씌웁니다. PyTorch 의 `F.scaled_dot_product_attention(..., is_causal=True)` 한 옵션이면 자동 적용됩니다.


````python
# is_causal=True 한 번 통과시켜 attention 행렬 시각화
# (직접 attn matrix 를 보려면 SDPA 대신 manual 버전 사용)
def visualize_causal_mask(L=8):
    mask = torch.tril(torch.ones(L, L))
    plt.figure(figsize=(4, 4))
    plt.imshow(mask, cmap="gray_r")
    plt.title(f"Causal mask (L={L})\n흰색 = 가려짐, 검정 = 볼 수 있음")
    plt.xlabel("key idx (보는 대상)"); plt.ylabel("query idx (보는 주체)")
    plt.show()

visualize_causal_mask(L=L)

# 실제 모델에 causal flag 만 켜면 끝
out_causal = model(token_ids, is_causal=True)
print("causal mode output shape :", out_causal.shape, "  (shape 은 동일, attention 패턴만 변경)")
````

## 9. (보너스) `nn.MultiheadAttention` 과 출력이 정말 같은가?

직접 만든 `MultiHeadSelfAttention` 이 PyTorch 의 표준 모듈과 수치적으로 일치하는지 확인합니다. 일치하면 우리가 짠 코드의 정확성이 검증된 것입니다.


````python
# 같은 가중치를 공유하도록 직접 만든 모듈의 weight 를 nn.MultiheadAttention 에 복사
my  = MultiHeadSelfAttention(dim=D, n_heads=H, bias=True).to(device).eval()
ref = nn.MultiheadAttention(embed_dim=D, num_heads=H, batch_first=True, bias=True).to(device).eval()

with torch.no_grad():
    # nn.MultiheadAttention 의 in_proj_weight = [W_q; W_k; W_v] (3D, D), in_proj_bias = [b_q; b_k; b_v]
    ref.in_proj_weight.copy_(my.qkv_proj.weight)
    ref.in_proj_bias.copy_(my.qkv_proj.bias)
    ref.out_proj.weight.copy_(my.out_proj.weight)
    ref.out_proj.bias.copy_(my.out_proj.bias)

x_test = torch.randn(N, L, D, device=device)

with torch.no_grad():
    my_out = my(x_test)
    ref_out, _ = ref(x_test, x_test, x_test, need_weights=False)

print("max abs diff :", (my_out - ref_out).abs().max().item())
print("close (atol=1e-5) ? :", torch.allclose(my_out, ref_out, atol=1e-5))
````

## 10. 실습 과제

### 과제 1. Toy 언어 모델로 학습시켜보기
`TinyEncoder` 의 마지막에 `nn.Linear(D, vocab_size)` 헤드를 추가해 next-token prediction 모델로 만들고, 짧은 문자열 데이터셋(예: shakespeare, "the quick brown fox..." 반복) 으로 100 epoch 학습. `is_causal=True` 로 학습해야 합니다 (decoder/GPT 모드).

### 과제 2. PE 끄고 학습 비교
`pos_enc` 를 identity 로 바꾸고 (PE 제거) 같은 데이터로 학습. **위치 정보 없이는 loss 가 거의 떨어지지 않는** 것을 확인하세요. ("Self-Attention 은 순서 정보가 없다" 는 슬라이드 #128 의 메시지를 직접 체험)

### 과제 3. Head 수를 바꾸며 ablation
`H = 1, 2, 4, 8` 로 훈련 → val loss 비교. head 가 너무 많으면(D/H 가 너무 작아지면) 표현력이 오히려 떨어질 수 있습니다.

### 과제 4. (advanced) 학습된 attention 행렬 시각화
어느 한 layer / head 의 `attention weight` 를 학습 후에 추출해 heatmap 으로 그려보세요. SDPA 가 fused kernel 이라 weight 가 직접 안 나오므로, `MultiHeadSelfAttention.forward` 에 `return_attn_weights=True` 옵션을 추가하고 manual SDPA 로 fallback 하는 코드를 직접 짜야 합니다.

### 과제 5. (advanced) NestedTensor / torch.compile() 적용
공식 튜토리얼 마지막 부분의 NestedTensor + `torch.compile(model)` 을 적용해 padding 없이 가변 길이 시퀀스를 처리하고 속도가 어떻게 변하는지 측정. 공식 튜토리얼 링크: https://docs.pytorch.org/tutorials/intermediate/transformer_building_blocks.html
