## reshape
- torch.view : original tensor의 memory 공유 (contiguity 제한)
- torch.reshape: contiguity 제한 없음. 필요에 따라 new tensor 생성. numpy 와 유사한 operation을 위해 기능 제공

```python
print(a)
print(a.view(5, 1))
print(a.reshape(5, 1))
```

```python
# 4x4 크기의 랜덤한 텐서 생성
x = torch.randn(4, 4)

# view: original tensor의 memory 공유 (contiguity 제한)
y = x.view(16)
# reshape: contiguity 제한 없음, -1은 나머지 차원의 크기를 자동 계산
z = x.reshape(-1, 8)

print(x.size(), y.size(), z.size())
```

```python
# transpose 후에는 contiguity가 깨질 수 있음
# reshape은 contiguity 제한이 없지만, view는 제한이 있음
x = torch.randn(4, 4)
x.transpose(0, 1).reshape(-1, 8)  # reshape은 가능
# x.transpose(0, 1).view(-1, 8)   # view는 에러 발생 가능
```

### torch.sum, torch.max, torch.min, torch.argmax, torch.argmin

- torch.sum(input, dim, keepdim=False, dtype=None) → Tensor  
주어진 차원 dim에서 입력 텐서의 각 행의 합(max, min)을 반환합니다.

- dim 을 지정한 경우, 지정된 dim을 축소하는 것입니다. 따라서 dim 0(행)을 접으면 하나의 행이 됩니다(열 단위로 합산).

```python
x = torch.tensor([
     [1, 2, 3],
     [4, 5, 6]
   ])
x.shape
```

```python
torch.sum(x)
```

```python
torch.sum(x, dim=0)
```

```python
torch.sum(x, dim=1)
```

```python
torch.max(x, dim=0), torch.min(x, dim=0)
```

```python
torch.max(x, dim=1), torch.min(x, dim=1)
```

```python
torch.argmax(x), torch.argmin(x)
```

```python
torch.argmax(x, dim=0), torch.argmin(x, dim=0)
```

```python
torch.argmax(x, dim=1), torch.argmin(x, dim=1)
```
